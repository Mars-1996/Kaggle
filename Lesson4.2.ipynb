```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
candy_filepath = "candy.txt"

# Fill in the line below to read the file into a variable candy_data
candy_data = pd.read_csv(candy_filepath, index_col = "id")

#candy_data.head() to show the first 5 rows
candy_data[0:11]
#to show the indicated numbers of rows
```

# Do people prefer candies with high sugar content?

```python
# Scatter plot showing the relationship between 'sugarpercent' and 'winpercent'
sns.scatterplot(x = candy_data["sugarpercent"], y = candy_data["winpercent"])
```

```python
# Scatter plot w/ regression line showing the relationship between 'sugarpercent' and 'winpercent'
sns.regplot(x = candy_data["sugarpercent"], y = candy_data["winpercent"])
```

The first plot does not show any correlation between the sugar content and the preferences registered with the survey. The addition of a regression line though shows a slight correlation, indicating a preference towards candies with high sugar content

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
sns.scatterplot(x = candy_data["pricepercent"], y = candy_data["winpercent"], hue = candy_data["chocolate"])

# Color-coded scatter plot w/ regression lines
sns.lmplot(x = "pricepercent", y = "winpercent", hue = "chocolate", data = candy_data) 
```

The blue regression line (with chocolate) has a positive slope, indicating that more expensive candies with chocolate are prefered.
Since the regression line for candies without chocolate has a negative slope, we can say that if candies don't contain chocolate, they tend to be more popular when they are cheaper

```python
# Scatter plot showing the relationship between 'chocolate' and 'winpercent'
sns.swarmplot(x = candy_data["chocolate"], y = candy_data["winpercent"], s = 5)
```
