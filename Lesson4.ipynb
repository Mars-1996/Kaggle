```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
insurance_filepath = "insurance.txt"

insurance_data = pd.read_csv(insurance_filepath, index_col = "age")

insurance_data.head()
```

```python
plt.figure(figsize=(8,6))


sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
#the plot suggests that the body mass index (bmi) and the charges are positively correlated. 
#This interpretation makes sense if we consider that higher bmi is associated to a highe risk of diseases.
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
#with the regplot function I can add a regression line in the plot.
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
# In this graph I add a third variable (smoker yes/no) that I differentiate with the two colors.

sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
#This plot adds two regression lines, for smokers and non smokers.

#This shows that smokers pay more their insurance if compared to non smokers.
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'], s = 4)
#with s = 4 I define the size of each dot in the graph. If the dots are too big they do not all fit the graph.
```
